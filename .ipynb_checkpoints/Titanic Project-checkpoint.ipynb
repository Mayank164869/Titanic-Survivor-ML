```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import shap
```

# Data Collection & Processing

```python
# load the data from csv file to Pandas DataFrame
titanic_data = pd.read_csv("C:/Users/mayan\Python Projects/Titanic Survivor Project//train.csv",encoding="charmap")
```

```python
# printing the first 5 rows of the dataframe
titanic_data.head()
```

```python
# number of rows and Columns
titanic_data.shape
```

```python
# check the number of missing values in each column
titanic_data.isnull().sum()
```

# Handling the Missing values

```python
# drop the "Cabin" column from the dataframe
titanic_data = titanic_data.drop(columns='Cabin', axis=1)
```

```python
# replacing the missing values in "Age" column with mean value
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
```

```python
# finding the mode value of "Embarked" column
print(titanic_data['Embarked'].mode())
```

```python
print(titanic_data['Embarked'].mode()[0])
```

```python
# replacing the missing values in "Embarked" column with mode value
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
```

```python
# check the number of missing values in each column
titanic_data.isnull().sum()
```

# Encoding the Categorical Columns


```python
titanic_data['Sex'].value_counts()
```

```python
titanic_data['Embarked'].value_counts()
```

```python
# converting categorical Columns
titanic_data.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
titanic_data.head()
```

# Separating features & Target

```python
X = titanic_data.drop(columns = ['PassengerId','Name','Ticket','Survived'],axis=1)
Y = titanic_data['Survived']
```

```python
print(X)
```

```python
print(Y)
```

#  Splitting the data into training data & Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training

# 1. Logistic Regression

```python
lr = LogisticRegression()
# training the Logistic Regression model with training data
lr.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = lr.predict(X_train)
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = lr.predict(X_test)
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```

```python
explainer = shap.LinearExplainer(lr, X_test)
shap_values = explainer.shap_values(X_test)
#The summary plot
shap.summary_plot(shap_values, X_test)
```

```python
shap.initjs()
#The individual force plot
shap.force_plot(explainer.expected_value, shap_values[0,:], X.iloc[0,:])
```

```python
#The Collective force plot
shap.force_plot(explainer.expected_value, shap_values, X_test)
```

```python
#The Dependence force plot
shap.dependence_plot("Age", shap_values, X_test)
```

# 2. K Nearest Neighbour

```python
knn = KNeighborsClassifier()
# training the K Nearest Neighbour model with training data
knn.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = knn.predict(X_train)
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = knn.predict(X_test)
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```

```python
knn_explainer = shap.KernelExplainer(knn.predict,X_test)
knn_shap_values = knn_explainer.shap_values(X_test)
```

```python
#The summary plot
shap.summary_plot(knn_shap_values, X_test)
```

```python
#The Dependence force plot
shap.dependence_plot("Age", knn_shap_values, X_test)
```

```python
#The individual force plot
shap.force_plot(knn_explainer.expected_value,knn_shap_values[1,:], X_test.iloc[1,:])
```

```python
#The Collective force plot
shap.force_plot(knn_explainer.expected_value, knn_shap_values, X_test)
```

# 3. Support Vector classifier

```python
svc = SVC(probability = True)
# training the Support Vector classifier model with training data
svc.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = svc.predict(X_train)
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = svc.predict(X_test)
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```

```python
svc_explainer = shap.KernelExplainer(svc.predict,X_test)
svc_shap_values = svc_explainer.shap_values(X_test)
```

```python
#summary plot
shap.summary_plot(svc_shap_values, X_test)
```

```python
#Dependence plot
shap.dependence_plot("Age", svc_shap_values, X_test)
```

```python
#Collective force plot
shap.force_plot(svc_explainer.expected_value, svc_shap_values, X_test)
```
